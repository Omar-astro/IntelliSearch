```python
import requests
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0"}
URL = "https://towardsdatascience.com/category/data-science/"
document_links = []

amount_of_articles = 70 # 10 divisibles only

for i in range(int(amount_of_articles / 10)):
    r = requests.get(URL, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    Links = soup.find_all('h2', class_="has-link-color")

    for h2 in Links:
        a_tag = h2.find('a')
        if a_tag:
            href = a_tag['href']
            document_links.append(href)

            URL = soup.find('a', class_="wp-block-query-pagination-next")['href']
document_links
```

```python
len(document_links)
```

```python
import pandas as pd

df = pd.DataFrame(columns=['index', 'Title', 'SubTitle', 'Author', 'Text', 'date','image', 'Link'])

for link in document_links:
    c1 = 1
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, 'html.parser')
    index = c1
    title = soup.find('h1', class_="wp-block-post-title").text.strip()
    Subtitle = soup.find('div', class_="tds-theme-post-subheading").find('p').text.strip()
    Author = soup.find('a', class_="wp-block-post-author-name__link").text.strip()
    date = soup.find("div", class_="wp-block-post-date").find('time').text.strip()
    image = soup.find("img", class_="attachment-post-thumbnail")['src']
    link_art = link
    Text = soup.find("div", class_="entry-content").text.strip()
    df.loc[len(df)] = [index, title, Subtitle, Author, Text, date, image, link]
    c1 += 1

df.head()
```

```python
import os

if os.path.exists('data/Articles.csv'):
    os.remove('data/Articles.csv')

df.to_csv('data/Articles.csv', index=False)
print(f"Saved new {'data/Articles.csv'}")
```
