# Fetching the Data

```python
import pandas as pd

df = pd.read_csv('data/bbc-news-summary.csv')
df.head()
```

# Cleaning the data

```python
before_preprocess = df['Articles'][0]
before_preprocess
```

```python
df['Articles'][0]
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

<h2> Word Lowering </h2>

```python
df['Articles'] = df['Articles'].str.lower()
df['Summaries'] = df['Summaries'].str.lower()
```

<h2> Tokenization </h2>

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
df['Articles'] = [word_tokenize(x) for x in df['Articles']]
```

<h2>Part of Speech (POS)</h2>

```python
from nltk import pos_tag
from nltk.tag import DefaultTagger

# nltk.download('averaged_perceptron_tagger_eng')

df["ArticlesPOS"] = [pos_tag(x) for x in df['Articles']]
df.head()
```

<h2> Stopword removal

```python
# from nltk.corpus import stopwords

# nltk.download('stopwords')
```

```python
mystopwords= stopwords.words("english")

df['Articles'] = [[word for word in tokens if word not in mystopwords] for tokens in df['Articles']]
df['Articles'][0]
```

<h2> Lemmatization </h2>

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

lemmatized_articles = []
for tokens in df["Articles"]:
    lemmatized_tokens = []
    for word in tokens:
        word = wnl.lemmatize(word, "n")
        word = wnl.lemmatize(word, "v")
        word = wnl.lemmatize(word, "a")
        lemmatized_tokens.append(word)
    lemmatized_articles.append(lemmatized_tokens)

df['Articles'] = lemmatized_articles
df['Articles'] = [" ".join(x) for x in df['Articles']]
df['Articles'][0]
```
