# Fetching the Data

```python
import pyterrier as pt
```

```python
import pandas as pd

df = pd.read_csv('data/Articles.csv')
df.head()
```

# Cleaning the data

```python
before_preprocess = df['Text'][0]
before_preprocess
```

```python
df['Text'][0]
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

<h2> Word Lowering </h2>

```python
df['Text'] = df['Text'].str.lower()
df['Title'] = df['Title'].str.lower()
df['SubTitle'] = df['SubTitle'].str.lower()
```

<h2> Tokenization </h2>

```python
# import nltk
# nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
df['Text'] = [word_tokenize(x) for x in df['Text']]
df['Title'] = [word_tokenize(x) for x in df['Title']]
df['SubTitle'] = [word_tokenize(x) for x in df['SubTitle']]
```

<h2>Part of Speech (POS)</h2>

```python
from nltk import pos_tag
from nltk.tag import DefaultTagger

# nltk.download('averaged_perceptron_tagger_eng')

df["TextPOS"] = [pos_tag(x) for x in df['Text']]
df.head()
```

<h2> Stopword removal

```python
from nltk.corpus import stopwords

# nltk.download('stopwords')
```

```python
mystopwords= stopwords.words("english")

df['Text'] = [[word for word in tokens if word not in mystopwords] for tokens in df['Text']]
df['Title'] = [[word for word in tokens if word not in mystopwords] for tokens in df['Title']]
df['SubTitle'] = [[word for word in tokens if word not in mystopwords] for tokens in df['SubTitle']]
df['Title'][0]
```

<h2> Lemmatization </h2>

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

lemmatized_articles = []
for tokens in df["Text"]:
    lemmatized_tokens = []
    for word in tokens:
        word = wnl.lemmatize(word, "n")
        word = wnl.lemmatize(word, "v")
        word = wnl.lemmatize(word, "a")
        lemmatized_tokens.append(word)
    lemmatized_articles.append(lemmatized_tokens)

df['Text'] = lemmatized_articles
df['Text'] = [" ".join(x) for x in df['Text']]

#-----------------------

wnl = WordNetLemmatizer()

lemmatized_articles = []
for tokens in df["Title"]:
    lemmatized_tokens = []
    for word in tokens:
        word = wnl.lemmatize(word, "n")
        word = wnl.lemmatize(word, "v")
        word = wnl.lemmatize(word, "a")
        lemmatized_tokens.append(word)
    lemmatized_articles.append(lemmatized_tokens)

df['Title'] = lemmatized_articles
df['Title'] = [" ".join(x) for x in df['Title']]

#----------------------------

wnl = WordNetLemmatizer()

lemmatized_articles = []
for tokens in df["SubTitle"]:
    lemmatized_tokens = []
    for word in tokens:
        word = wnl.lemmatize(word, "n")
        word = wnl.lemmatize(word, "v")
        word = wnl.lemmatize(word, "a")
        lemmatized_tokens.append(word)
    lemmatized_articles.append(lemmatized_tokens)

df['SubTitle'] = lemmatized_articles
df['SubTitle'] = [" ".join(x) for x in df['SubTitle']]
```

```python
import re

word_cont = {}
for sentence in df['Text']:
    tokens = word_tokenize(sentence)
    for i in range(len(tokens)-1):
        word = tokens[i]
        word_after = tokens[i+1]
        
        # Check both the word and the word immediately after
        if re.match(r'^\w+$', word) and re.match(r'^\w+$', word_after):
            if word in word_cont:
                if word_after in word_cont[word]:
                    word_cont[word][word_after] += 1
                else:
                    word_cont[word][word_after] = 1
            else:
                word_cont[word] = {}
                word_cont[word][word_after] = 1
```

```python
query = "Pandas"   

print("Your query:", query)

results = bm25.search(query)
print(results)
```

```python
rm3 = pt.rewrite.RM3(index_ref, fb_terms=10, fb_docs=100)
rm3_qe = bm25 >> rm3
expanded_query = rm3_qe.search(query).iloc[0]["query"]
expanded_query_formatted = ' '.join(expanded_query.split()[1:])
results_wqe = bm25.search(expanded_query_formatted)
print("   Before Expansion    After Expansion")
print(pd.concat([results[['docid','score']][0:5].add_suffix('_1'),
            results_wqe[['docid','score']][0:5].add_suffix('_2')], axis=1).fillna(''))
```

```python
# Inverted Index: term -> {doc_id: frequency}
inverted_index = {}

for doc in docs:
    doc_id = int(doc["docno"])
    tokens = doc["text"].split() 
    
    for word in tokens:
        if word not in inverted_index:
            inverted_index[word] = {}
        
        if doc_id not in inverted_index[word]:
            inverted_index[word][doc_id] = 0
        
        inverted_index[word][doc_id] += 1
```

```python
count = 0
for term in inverted_index:
    print(term, "->", inverted_index[term])
    count += 1
    if count == len(query):
        break
```
